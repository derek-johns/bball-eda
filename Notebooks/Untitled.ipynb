```python
import pandas as pd
```

```python
def load_data(year):
    url = "https://www.basketball-reference.com/leagues/NBA_" + str(year) + "_per_game.html"
    html = pd.read_html(url, header=0)
    print(html)
    df = html[0]
    print(df)
    raw = df.drop(df[df.Age == 'Age'].index)
    print(raw)
    raw = raw.fillna(0)
    print(raw)
    playerstats = raw.drop(['Rk'], axis=1)
    print(playerstats)
```

```python
load_data(2020)
```
